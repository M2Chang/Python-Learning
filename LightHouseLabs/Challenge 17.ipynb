```python
#Bar Plots and Specific Data

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("books.csv")

df.head(2)
```

```python
#What are the top 5 rated books in the dataset?

top_rated = df.sort_values(by='ratings_count', ascending = [False]).head(5)
top_rated
```

```python
plt.figure() #The Frame: We start our plot with a figure
plt.barh(y = top_rated['title'], width = top_rated['ratings_count']) #The Body: Declaring the specific bar plot statment
plt.title("Top Rated") #Stylistic Features: Adding the title
plt.show()
```

```python
#What are the top 5 books with the highest average rating?

```

```python
topfilter = df['ratings_count'] > df['ratings_count'].mean()
df_filter = df[topfilter]
tophighestrate = df_filter.sort_values(by='average_rating', ascending = [False]).head(5)
tophighestrate
```

```python
plt.figure() #The Frame: We start our plot with a figure
plt.barh(y = tophighestrate['title'], width = tophighestrate['average_rating']) #The Body: Declaring the specific bar plot statment
plt.title("Top Rated") #Stylistic Features: Adding the title
plt.show()
```
