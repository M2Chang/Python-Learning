```python
#sorting with pandas

import pandas as pd
wine_df = pd.read_csv('winequality-red.csv')
wine_df
```

```python
#Which wines had a quality of 8 or higher and a residual sugar level above 5?

qualityfilter = wine_df['quality'] >= 8
sugarlevelfilter = wine_df['residual sugar'] >= 5

filtered_df = wine_df[qualityfilter | sugarlevelfilter]
```

```python
filtered_df.sort_values(by=['quality','residual sugar'], ascending = False)
```

```python
#How many wines in total had a quality of 8 and 7 and a citric acid level below 0.4?

qualityfilter2 = wine_df['quality'] == 7
qualityfilter3 = wine_df['quality'] == 8
acidfilter = wine_df['citric acid'] < 0.4
```

```python
filtered_df2 = wine_df[(qualityfilter2 | qualityfilter3) & acidfilter]
```

```python
filtered_df2['quality'].value_counts()
```
